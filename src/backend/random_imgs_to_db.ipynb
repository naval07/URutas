## Inclusión de imágenes en la base de datos

```python
import pandas as pd
import random
```

```python
# importación del dataset
cursos = pd.read_csv('Final_Cursos_normalizados.csv')
cursos.head(10)
```

```python
# Creación de la lista que contiene imágenes aleatorias
num_cursos = cursos.shape[0]
num_imgs = 10
lista_imgs = []

for i in range(num_cursos):
    numero_random = random.randrange(1,num_imgs+1)
    ruta = f'/fondoCartas/img{numero_random}.jpg'
    lista_imgs.append(ruta)
```

```python
# Creación de la tabla nueva
cursos['ruta_img'] = lista_imgs
cursos.head(5)
```

```python
# Guardado de nueva db
cursos.to_csv('Final_Cursos_normalizados.csv')
```

## Creación de nuevo catálogo para la página de inicio

```python
import json
```

```python
cursos = pd.read_csv('Final_Cursos_normalizados.csv')
```

```python
# Lista con los cursos en formato diccionario (json)
lista_cursos = []
for i in range(cursos.shape[0]):    
    # Creación del formato
    curso_dict = {
        "index": int(cursos.loc[i][0]),
        "Título": cursos.loc[i][1],
        "proposito": cursos.loc[i][4],
        "area": cursos.loc[i][3],
        "rutaImg": cursos.loc[i][9]
    }
    lista_cursos.append(curso_dict)
```

```python
# Escritura del archivo json
ruta_catalogo = "../components/catalogo/catalogo.json"
open(ruta_catalogo, 'w').close() # Se borra el archivo por completo para evitar acumulación en caso de correrse varias veces
with open(ruta_catalogo, "w", encoding='utf-8') as json_file:
    json.dump(lista_cursos, json_file, ensure_ascii=False, indent=2)
```
